# face detection model

```python
face_detection_model = './models/openface/res10_300x300_ssd_iter_140000.caffemodel'
face_detection_proto = './models/openface/deploy.prototxt.txt'
face_descriptor = './models/openFace/openface.nn4.small2.v1.t7'

detector_model = cv2.dnn.readNetFromCaffe(face_detection_proto,face_detection_model)
descriptor_model = cv2.dnn.readNetFromTorch(face_descriptor)
```

```python
# step 1: face detection
def proccesImageData(image_path, display_name):
    img = cv2.imread(image_path)
    image = img.copy()
    h,w = image.shape[:2]
    img_blob = cv2.dnn.blobFromImage(image, 1, (300, 300), (104, 177, 123), swapRB=False, crop=False)

    # set input
    detector_model.setInput(img_blob)
    detections = detector_model.forward()

    if len(detections) > 0:
        i = np.argmax(detections[0,0,:,2])# consider the face with max confidence score
        confidence = detections[0,0,i,2]
        if confidence > 0.50:
            box = detections[0,0,i,3:7]*np.array([w,h,w,h])
            (startx,starty,endx,endy) = box.astype('int')
            # step-2: Feature Extraction or Embedding
            roi = image[starty:endy,startx:endx].copy()
            # get the face descriptors
            faceblob = cv2.dnn.blobFromImage(roi,1/255,(96,96),(0,0,0),swapRB=True,crop=True)
            descriptor_model.setInput(faceblob)
            vectors = descriptor_model.forward()
            
            return vectors
        
        return None
```

```python
proccesImageData('./test/Training Data/Sachin Tendulkar/2200.jpg', 'Sachin Tendulkar')
```

```python
data = dict(data=[], label=[])
target_suffix = './test/Training Data/'
folders = os.listdir(target_suffix)
folders = list(filter(lambda input: input.find('.jpg') == -1 and input.find('.png') == -1 and input.find('.DS_Store'), folders))
for folder in folders:
    filenames = os.listdir(f'{target_suffix}{folder}')
    filenames = list(filter(lambda input: input.find('.j') != -1 or input.find('.png') != -1, filenames))
    for filename in filenames:
        try:
            vector = proccesImageData(f'{target_suffix}{folder}/{filename}', folder)
            if vector is not None:
                data['data'].append(vector)
                data['label'].append(folder)
                print("Feature extracted!")
                
        except:
            print(f"{filename} Failed to extract!")
            pass
            
```

```python
# save data 
pickle.dump(data,open('data_face_features.pickle', mode='wb'))
#later feed saved data to the model
```
